## 箱型图

```python
def tidai(x):
    if x=="居住型":
        y="Residential"
    if x=="混合型":
        y="Mixed"
    if x=="商服型":
        y="Commercial"
    if x=="产业型":
        y="Industrial"
    return y
    
```

```python
def rain(x):
    if x=="小雨":
        y="Light rain"
    if x=="中雨":
        y="Medium Rain"
    if x=="大雨":
        y="Heavy rain"
    if x=="暴雨":
        y="Rainstorm"
    return y
```

```python
cc=pd.read_clipboard()
cc.Occat=cc.apply(lambda x:tidai(x.Occat),1)
cc.rain=cc.apply(lambda x:rain(x.rain),1)
cc
```

```python

fig,ax=plt.subplots(figsize=(6.96,3.66),dpi=150)
# for i in np.arange(-0.2,0.2,0.05):
#     ax.axhline(i, color="grey", ls='--',clip_on=False)
sns.boxplot(x='Occat',y='change',hue='rain',data=cc)
plt.ylabel('Passenger flow volatility',fontsize=12)
plt.xlabel(None)
# ax.tick_params(axis="x", bottom=False)#隐藏刻度线
for spine in ['top','right']:
    ax.spines[spine].set_color('none')
ax.set_yticks(np.arange(-0.3,0.25,0.05))
ax.set_ylim(-0.35,0.22)
ax.legend(loc=2,bbox_to_anchor=(1.05,1.0),fontsize=10,borderaxespad=0.,edgecolor='black')
plt.savefig(r'C:\Users\飘哥\Desktop\分箱图晚高峰.png',dpi=300,bbox_inches='tight')
```

```python
#求箱型图的上须和下须

def outliers_proc(data, col_name, scale):
    """
    用于清洗异常值，默认box_plot(scale=3)进行清洗
    param data: 接收pandas数据格式
    param col_name: pandas列名
    param scale: 尺度
    """
        
    def box_plot_outliers(data_ser, box_scale):
        """
        利用箱线图去除异常值
        :param data_ser: 接收 pandas.Series 数据格式
        :param box_scale: 箱线图尺度
        """
        iqr = box_scale * (data_ser.quantile(0.75) - data_ser.quantile(0.25))
        val_low = data_ser.quantile(0.25) - iqr
        val_up = data_ser.quantile(0.75) + iqr
        rule_low = (data_ser < val_low)
        rule_up = (data_ser > val_up)
        return (rule_low,rule_up),(val_low,val_up)
    
    data_n = data.copy()
    data_serier = data_n[col_name]
    rule, value = box_plot_outliers(data_serier,box_scale=scale)
    index = np.arange(data_serier.shape[0])[rule[0]|rule[1]]
#     print("Delete number is:{}".format(len(index)))
    data_n = data_n.drop(index)
    data_n.reset_index(drop=True, inplace=True)
#     print("Now column number is:{}".format(data_n.shape[0]))
    index_low = np.arange(data_serier.shape[0])[rule[0]]
    outliers = data_serier.iloc[index_low]
#     print("Description of data less than the lower bound is:")
#     print(pd.Series(outliers).describe())
    index_up = np.arange(data_serier.shape[0])[rule[1]]
    outliers = data_serier.iloc[index_up]
#     print("Description of data larger than the upper bound is:")
#     print(pd.Series(outliers).describe())
    
#     fig, ax = plt.subplots(1,2, figsize=(10,7))
#     sns.boxplot(y=data[col_name],data=data,palette="Set1",ax=ax[0])
#     sns.boxplot(y=data_n[col_name],data=data_n,palette="Set1",ax=ax[1])
    return data_n
```
